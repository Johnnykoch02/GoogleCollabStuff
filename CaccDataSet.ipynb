```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_excel("CaccDataSet.xlsx")
x = df['Corn Price']
y = df['DE Price']

np.corrcoef(x, y)
plt.scatter(x, y)
plt.title('Corn Price vs. DE Price')
eq = np.poly1d(np.polyfit(x, y, 1))
plt.plot(np.unique(x), np.poly1d(np.polyfit(x, y, 1))(np.unique(x)), color='red')
plt.show()


my_fitting = np.polyfit(x, y, 1, full=True)
coeff = my_fitting[0]

### Residual or Sum of Square Error (SSE)
SSE = my_fitting[1][0]

### Determining the Sum of Square Total (SST)
## the squared differences between the observed dependent variable and its mean
diff = y - y.mean()
square_diff = diff ** 2
SST = square_diff.sum()

###  Now getting the coefficient of determination (R2)
R2 = 1 - SSE/SST 

print('R Squared: ', R2)


cor_df = df.dropna(subset=['Corn Positive ', 'DE resective '])

x = cor_df['Corn Positive ']
y = cor_df['DE resective ']

np.corrcoef(x, y)
plt.scatter(x, y)
plt.title('Corn Positive % vs. DE % Change')
plt.plot(np.unique(x), np.poly1d(np.polyfit(x, y, 1))(np.unique(x)), color='red')
plt.show()



my_fitting = np.polyfit(x, y, 1, full=True)
coeff = my_fitting[0]

### Residual or Sum of Square Error (SSE)
SSE = my_fitting[1][0]

### Determining the Sum of Square Total (SST)
## the squared differences between the observed dependent variable and its mean
diff = y - y.mean()
square_diff = diff ** 2
SST = square_diff.sum()

###  Now getting the coefficient of determination (R2)
R2 = 1 - SSE/SST 

print('R Squared: ', R2)

```
